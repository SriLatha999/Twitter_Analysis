```python
def cleanTxt(text):
    text= re.sub('@[A-Za-z0-9]+','',text ) #removing mentions
    text= re.sub("#",'',text) #removing #
    text= re.sub('RT[\s]+','',text) # removing Retweets
    text= re.sub('https?:\/\/\S+','',text) #removing links
    return text
```

```python
df['Text']= df['Text'].apply(cleanTxt)
```

```python
df
```

```python
df.drop(['Datetime','Tweet Id','Username'], axis=1)
```

```python
# create a function to get the subjectivity of all the tweets
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# create a function to get the Polarity of all the tweets
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

# create 2 columns 'Subjectivity' and 'Polarity'
df['Subjectivity']= df['Text'].apply(getSubjectivity)
df['Polarity']= df['Text'].apply(getPolarity)
```

```python
STOPWORDS = set(stopwords.words('english'))
def cleaning_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
df['Text']=df['Text'].apply(lambda text: cleaning_stopwords(text))
df['Text'].head()
```

```python
english_punctuation = string.punctuation
punctuations_list= english_punctuation
def cleaning_punctuations(text):
    translator = str.maketrans('','',punctuations_list)
    return text.translate(translator)
```

```python
df['Text']=df['Text'].apply(lambda x: cleaning_punctuations(x))
df['Text'].tail()
```

```python
def cleaning_repeating_char(text):
    return re.sub(r'(.)\1+', r'\1', text)
```

```python
df['Text']=df['Text'].apply(lambda x: cleaning_repeating_char(x))
df['Text'].tail()
```

```python
def cleaning_URLs(data):
    return re.sub('((www\.[^\s]+)|(https?://[^\s)+))','', data)
```

```python
def cleaning_numbers(data):
    return re.sub('[0-9]+','', data)
```

```python
df['Text']=df['Text'].apply(lambda x: cleaning_numbers(x))
df['Text'].tail()
```

```python
tokenizer = RegexpTokenizer(r'\w+')
df['Text']= df['Text'].apply(tokenizer.tokenize)
```

```python
df['Text'].head()
```

```python
#stemming
```

```python
st = nltk.PorterStemmer()
def stemming_on_text(data):
    text = [st.stem(word) for word in data]
    return data

df['Text']=df['Text'].apply(lambda x: stemming_on_text(x))
```

```python
df['Text'].head()
```

```python
#lemmatizer
```

```python
lm = nltk.WordNetLemmatizer()
def lemmatizer_on_text(data):
    text = [lm.lemmatize(word) for word in data]
    return data

df['Text']=df['Text'].apply(lambda x: lemmatizer_on_text(x))
```
